```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = sns.load_dataset('diamonds')
df.head()
```

```python
print(df.describe())
print(df.info())
print(df.isnull().sum())
```

```python
plt.figure(figsize=(10, 6))
numeric_df = df.select_dtypes(include='number')  # Select numeric columns only
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(df['price'], bins=50, kde=True)
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.boxplot(x='cut', y='price', data=df)
plt.title('Price by Cut')
plt.show()
```

```python
fig = px.scatter(df, x='carat', y='price', color='cut', title='Carat vs Price by Cut')
fig.show()
```

```python
fig = px.histogram(df, x='price', color='cut', nbins=50, title='Price Distribution by Cut')
fig.show()
```
